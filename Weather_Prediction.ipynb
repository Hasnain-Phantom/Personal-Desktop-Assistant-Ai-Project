```python
import pandas as pd
```

```python
weather = pd.read_csv("/content/drive/MyDrive/local_weather.csv", index_col="DATE")
#reading the csv file in to a new variable
#use the data column as index
weather
```

```python
perce=(weather.apply(pd.isnull).sum()/weather.shape[0])*100
perce
#percentage of values in each column that are null
```

```python
data=weather[['PRCP','SNOW','SNWD','TMAX','TMIN']].copy()
#core columns for our weather prediction
```

```python
data.columns=['precipitation','snow','snow_depth','temp_max',"temp_min"]
#renaming columns
```

```python
del data['snow']
del data['snow_depth']
data
#removed snow and snow depth columns as they are mostly empty
```

```python
data[pd.isnull(data['precipitation'])]
```

```python
#Fill the null values with the average value
avg1 = data['precipitation'].mean()
avg2 = data['temp_max'].mean()
avg3 = data['temp_min'].mean()

data['precipitation'].fillna(avg1, inplace=True)
data['temp_max'].fillna(avg2, inplace=True)
data['temp_min'].fillna(avg3, inplace=True)
```

```python
(data.apply(pd.isnull).sum()/data.shape[0])*100
#all values have been filled
#there are no none values in any of the columns
```

```python
data['target']=data.shift(-1)['temp_max']
data=data.iloc[:-1,:].copy()
```

```python
from sklearn.linear_model import Ridge
reg=Ridge(alpha=0.1)
```

```python
predictors=['temp_max','temp_min','precipitation']
target_var=['target']
```

```python
train_set=data.loc[:"2020-12-31"] #using the old data to make predictions for the future
test_set=data.loc['2021-01-01':] #using the new data to test our model
```

```python
reg.fit(train_set[predictors],train_set['target']) #using predictors to predict the target variable
```

```python
predictions=reg.predict(test_set[predictors]) #using test_data to make predictions
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
error=mean_absolute_error(test_set['target'],predictions) #calculating the mean_abs_error between our predictions and the correct values
error
```

```python
analysis_table = pd.concat ([ test_set['target'],pd.Series(predictions,index=test_set.index)],axis=1)
analysis_table.columns=["target",'predictions']
analysis_table
#creating an analysis table to analyze each row
```

```python
analysis_table.plot()
```

```python
def predictions_generator(predictors,data,reg):
  train_set=data.loc[:"2020-12-31"] #using the old data to make predictions for the future
  test_set=data.loc['2021-01-01':] #using the new data to test our model
  reg.fit(train_set[predictors],train_set['target']) #using predictors to predict the target variable
  predictions=reg.predict(test_set[predictors]) #using test_data to make predictions
  error=mean_absolute_error(test_set['target'],predictions) #calculating the mean_abs_error between our predictions and the correct values
  analysis_table = pd.concat ([ test_set['target'],pd.Series(predictions,index=test_set.index)],axis=1)
  analysis_table.columns=["target",'predictions']
  analysis_table
  return error,analysis_table
```
